```python
import matplotlib.pyplot as plt
from tqdm import trange
from math import factorial
from collections import Counter

from utils import partitions
from searchspace import searchspace
```

```python
smart_vals = []
naive_vals = []
for i in trange(9):
    smart, naive = searchspace(i, i*i)
    smart_vals.append(smart)
    naive_vals.append(naive)
```

```python
figure, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=True)
figure.suptitle("Comparing the number of operations")

ax1.plot(smart_vals, '-go')
ax1.set_yscale('log') # cannot pass y_scale as a kwarg to plt.subplot, hence this thing here, ugh.
ax2.plot(naive_vals, '-bo')
figure.supxlabel("Grid Size")
figure.supylabel("Search Space")
```
